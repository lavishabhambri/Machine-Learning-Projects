```python
import numpy as np
from sklearn import datasets
cancer_ds = datasets.load_breast_cancer()
```

```python
cancer_ds
```

```python
x = cancer_ds.data
y = cancer_ds.target
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn import model_selection
x_train, x_test , y_train, y_test = model_selection.train_test_split(x, y,random_state = 0)
```

```python
#The data is not normalized, so do normalization
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#training data
train_ones = np.ones(x_train.shape[0]).reshape(-1, 1)
x_train= np.append(x_train, train_ones, axis=1)

#test data
test_ones = np.ones(x_test.shape[0]).reshape(-1, 1)
x_test = np.append(x_test, test_ones, axis=1)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
def step_gradient(X_train, Y_train, learning_rate, m,j):
    # Calculate new slope for jth feature
    m_j = 0
    n_data_pts = X_train.shape[0]
    N = len(m)
    for i in range(n_data_pts):
        # calculate the formula m1xi(1)+m2xi(2)+...
        x_i = X_train[i, :]
        y_i = Y_train[i]
        temp_sum = 0
        for k in range(N):
            temp_sum += m[k]*x_i[k]
            
        h = 1/(1+np.exp(-temp_sum))
        e = y_i - h
        #complete formula
        m_j += (-1/n_data_pts) * (e) * x_i[j]
    # update m[j] and return
    m[j] = m[j] - (learning_rate*m_j)
    return m[j]
```

```python
def gd(x_train, y_train, learning_rate, num_iterations):
    m = [0]* (x_train.shape[1])
    N = len(m)
    #put c=1
    m[-1] = 1
    
    for i in range(num_iterations):
        for j in range(N):
            m[j] = step_gradient(x_train, y_train, learning_rate, m , j)
        print(i, 'Cost : ', cost(x_train, y_train, m))
    return m
```

```python
def cost(X_train, Y_train, m):
    
        # This will calculate error
    cost = 0
    n_data_pts = len(X_train)
    N = len(m)
    for i in range(n_data_pts):
        x_i = X_train[i, :]
        y_i = Y_train[i]
        temp_sum = 0
        for k in range(N):
            temp_sum += m[k]*x_i[k]
        h = 1/(1+np.exp(-temp_sum))

        cost += (1/n_data_pts) * (y_i * np.log(h) + (1-y_i) * np.log(1-h))
    return cost
```

```python
def pred(x_test, m):
    pred = []
    n_data_pt = x_test.shape[0]
    n_features = x_test.shape[1]
    for i in range(n_data_pt):
        total_sum = 0
        for j in range(n_features):
            total_sum += m[j] * x_test[i][j]
            
        h = 1/(1+ np.exp(-total_sum))
        if h <0.5:
            pred.append(0)
        else:
            pred.append(1)
            
    return pred
```

```python
def accuracy(y_pred, y_true):
    return 100 * (np.mean(y_pred == y_true))
```

```python
learning_rate = 0.01
num_iterations = 1000
m = gd(x_train, y_train, learning_rate, num_iterations)
y_pred = pred(x_test, m)
print(accuracy(y_pred, y_test))
```
