# Data Analysis

### Deleting unnecessary columns

```python
del df1['Name']
del df1['Ticket']
del df1['Fare']
del df1['Cabin']

del df2['Name']
del df2['Ticket']
del df2['Fare']
del df2['Cabin']
```

```python
df1
```

```python
df2
```

### Handling String Values

```python
#on Sex column
def getGender(s):
    if s =='male':
        return 0
    else:
        return 1
df1['Sex'] = df1.Sex.apply(getGender)
print(df1)

df2['Sex'] = df2.Sex.apply(getGender)
print(df2)
```

```python
#on embarked column
def Embarked(s):
    if s=='C':
        return 0
    elif s=='Q':
        return 1
    else:
        return 2
    
df1['Embarked'] = df1.Embarked.apply(Embarked)
print(df1)

df2['Embarked'] = df2.Embarked.apply(Embarked)
print(df2)
```

```python
df1.isnull().sum()
```

### Handling the missing values in Age column

```python
#filling the nan values of age column of df1 by the mean of the persons survived and unsurvived respectively.

#suvived
s = df1[df1.Survived == 1].Age.mean()
df1["Age"]=np.where(pd.isnull(df1.Age) & df1["Survived"]==1  ,s, df1["Age"])

us=df1[df1.Survived == 0].Age.mean()
df1.Age.fillna(us,inplace=True)
df1
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
df2.isnull().sum()
```

### So there are some missing values in Age too. So apply same as above in Age column of df2.

```python
#filling the nan values of age column of df2 by the mean of the persons.

df2.Age.fillna(df2.Age.mean(),inplace=True)
df2
```

```python
df2.isnull().sum()
```

# Logistic Regression Code

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(C=0.1, max_iter=100, penalty='l1', solver='saga')
```

```python
train_data = df1.values
test_data = df2.values
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
x = train_data[:,0:6]
y = train_data[:,6:7]
```

```python
print(x.shape)
print(y.shape)
```

# First testing the accuracy on test data by splitting training data into training and test data

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
print(y_pred)
```

```python
score = lr.score(x_test, y_test)
print(score)
```
